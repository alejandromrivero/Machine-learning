```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

###datos del razon contable

```python
razao=pd.DataFrame(pd.read_excel('X:/fabiana 03-04-2020/razao01.xlsx'))
razao['Data']=pd.to_datetime(razao['Data'])
razao['Débito']=pd.to_numeric(razao['Débito'])
razao['Crédito']=pd.to_numeric(razao['Crédito'])
razao['Titulo']=pd.to_numeric(razao['Titulo'])
razao['Débito']=razao['Débito'].replace({np.nan:0.00})
razao['Crédito']=razao['Crédito'].replace({np.nan:0.00})
razao['Saldo']=razao['Crédito']-razao['Débito']
razao
```

```python
razao['Crédito'].sum()-razao['Débito'].sum()
```

###datos de los titulos

```python
titulo=pd.DataFrame(pd.read_excel('X:/fabiana 03-04-2020/titulos01.xlsx'))
titulo['Dt emissão']=pd.to_datetime(titulo['Dt emissão'])
titulo['Dt vecto']=pd.to_datetime(titulo['Dt vecto'])
titulo['Dt baixa']=pd.to_datetime(titulo['Dt baixa'])
titulo['Titulo']=pd.to_numeric(titulo['Titulo'])
titulo['Valor original']=pd.to_numeric(titulo['Valor original'])
titulo['Valor da baixa']=pd.to_numeric(titulo['Valor da baixa'])
```

```python
datos=pd.DataFrame(pd.merge(razao,titulo,on='Titulo'))
```

```python
datos.dtypes
```

```python
datos_fin=datos[['Data','Credor','Histórico','Débito','Crédito','Saldo','Titulo',
'Documento','Parcela']]
datos_fin=pd.DataFrame(datos_fin)
```

```python
datos_fin=datos_fin.set_index('Titulo')
```

```python
relatorio_01=pd.DataFrame(datos_fin.groupby(['Credor','Titulo'])[['Débito','Crédito','Saldo']].sum())
relatorio_01.to_excel('X:/fabiana 03-04-2020/relatorio_01.xlsx')
```

```python
relatorio_02=pd.DataFrame(datos_fin.groupby(['Credor'])[['Débito','Crédito','Saldo']].sum())
relatorio_02.to_excel('X:/fabiana 03-04-2020/relatorio_02.xlsx')
```

```python
relatorio_03=pd.DataFrame(datos_fin.groupby(['Credor','Titulo','Documento'])[['Débito','Crédito','Saldo']].sum())
relatorio_03.to_excel('X:/fabiana 03-04-2020/relatorio_03.xlsx')
```

```python
relatorio_04=pd.DataFrame(datos_fin.groupby(['Credor','Titulo','Data'])[['Débito','Crédito','Saldo']].sum())
relatorio_04.to_excel('X:/fabiana 03-04-2020/relatorio_04.xlsx')
```
