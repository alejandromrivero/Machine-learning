```python
ts1=pd.Series(np.random.randn(600), 
                  index=pd.date_range(end='2000-01-15', periods=600,freq='M'))
ts1
```

```python
ts1.resample('Q-JAN').sum()
```

```python
ts2=ts1.resample('A-DEC').sum()
ts2
```

```python
ts2.index
```

```python
sn.scatterplot(x=ts2.index,y=ts2, label='Serie', color='grey', 
               marker='s',edgecolor='blue')
plt.xlim('1949-01-01','2001-01-01')

plt.grid(True,linestyle='dashed', color='blue')
plt.title('Serie Temporal', size=18, color='blue')
plt.xlabel('Quatrimestre/Ano', size=14, color='blue')
plt.ylabel('valor', size=14, color='blue')
plt.legend(loc='best')
plt.savefig('F:/scatter_ts1.png')
```
