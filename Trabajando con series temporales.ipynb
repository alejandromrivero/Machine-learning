```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
valor=pd.DataFrame(np.random.randint(2,50,size=5000),columns=['Valor'], index=np.arange(5000))
valor.index.names=['indice']
valor
```

```python
valor1=pd.DataFrame(np.random.rand(5000)+np.random.randint(2,8), columns=['Valor1'], index=np.arange(5000))
valor1.index.names=['indice']
valor1
```

```python
data=pd.DataFrame(pd.date_range(start='2020-01-01',periods=5000, freq='D'),columns=['Data'],index=np.arange(5000))
data.index.names=['indice']
data
```

```python
valor2=pd.DataFrame(np.random.randint(2,150,size=5000),columns=['Valor2'], index=np.arange(5000))
valor2.index.names=['indice']
valor2
```

```python
df1=pd.merge(data, valor, on='indice')
df1
```

```python
df2=pd.DataFrame(pd.merge(df1,valor1, on='indice'))
df3=pd.DataFrame(pd.merge(df2,valor2, on='indice'))
df3
```

```python
df3.set_index('Data')
```

```python
df3.dtypes
```

```python
df3['Data']=pd.DatetimeIndex(df2['Data'])
```

```python
df3.dtypes
```

```python
df3=df3.set_index('Data')
```

```python
df4=pd.DataFrame(df3.resample('Y').mean())
df4.corr()
```

```python
df4
```

```python
plt.plot(df4.index,df4['Valor'], color='grey', linewidth=1.9)
plt.xticks(df4.index, rotation=60, color='blue')
plt.yticks(df4['Valor'], color='blue', rotation=0, fontsize=10)
plt.xlabel('Ano', color='red',fontsize=18)
plt.ylabel('Valores arrecadados (milhÃµes)', fontsize=18, color='blue')
plt.grid(True, linestyle='--', color='blue', linewidth=0.5)
```

```python
sb.scatterplot(df4.index,df4['Valor'], label='Valor arrecadado no periodo', color='red', markers='*')
plt.xlim('2019-12-31','2034-12-31')
plt.xlabel('Ano', color='red',fontsize=18)
plt.ylabel('Valores arrecadados (milhÃµes)', fontsize=18, color='blue')
plt.xticks(df4.index, rotation=90, color='blue')
#plt.yticks(df4['Valor'], color='blue', rotation=0, fontsize=10)
plt.grid(True, linestyle='--', color='blue', linewidth=0.5)
plt.legend(loc='best', fontsize=12)
```
