```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from statsmodels.stats.tests.test_multi import NA
from numpy.random import randn
```

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.plot(randn(1000).cumsum(),'g',label='one')
ax.plot(randn(1000).cumsum(),'b--',label='two')
ax.plot(randn(1000).cumsum(),'r.',label='three')
ax.legend(loc='best')
ax.set_xlabel('dias transcurridos')
ax.set_xlim([-100,1400])
ax.set_ylabel('Quantidade')
ax.set_ylim([-100,100])
ax.text(400,-35,'aqui va el texto',fontsize=10)
```
