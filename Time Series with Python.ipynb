###Filling Missing Values

```python
df_comp['spx']=df_comp['spx'].fillna(method='ffill')
df_comp['dax']=df_comp['dax'].fillna(df_comp['dax'].mean())
df_comp['ftse']=df_comp['ftse'].fillna(method='bfill')
df_comp['nikkei']=df_comp['nikkei'].fillna(df_comp['nikkei'].mean())
df_comp.isna().sum()
```

### White Noise

####White noise es utilizado para saber si a serie temporal tiene un padrao aleatorio, caso tenga un padrao aleatorio entonces no es posible hacer pronosticos con ella.

```python
wn=np.random.normal(loc=df['market value'].mean(), scale=df['market value'].std(), size=len(df))
```

```python
df['wn']=wn
```

```python
df.describe()
```

```python
plt.figure(figsize=(20,5))
plt.plot(df['wn'])
plt.title('White Noise Time Series', size=24)
plt.savefig('F:/WhiteNoise.jpg')
plt.show()
```

```python
rw=pd.read_csv('F:/Time series com Python/Time Series Analysis in Python 2020/05 Working with Time Series in Python/RandWalk.csv')
```

```python
rw['date']=pd.to_datetime(rw['date'], dayfirst=True)
rw.set_index('date', inplace=True)
rw.asfreq(freq='b')
```

```python
df['rw']=rw['price']
df.head()
```
