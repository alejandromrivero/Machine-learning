```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

#CMA

####Carregando os dados referentes ao razão contábil.

```
razao=pd.DataFrame(pd.read_excel('F:/Conciliacao/CMA/RAZAO_CMA.xls'))
razao.describe()
```

```
razao['Titulo']=pd.to_numeric(razao['Titulo']) #formateando como numero los titulos, debito, credito
razao['Débito']=pd.to_numeric(razao['Débito'],errors='coerce')
razao['Crédito']=pd.to_numeric(razao['Crédito'],errors='coerce')
razao['Titulo']=pd.to_numeric(razao['Titulo'])
razao['Débito']=razao['Débito'].replace({np.nan:0.00})
razao['Crédito']=razao['Crédito'].replace({np.nan:0.00})
razao['Saldo']=razao['Crédito']-razao['Débito'] #calculando los saldos por cada movimento contabil de cada un de los titulos
razao.set_index('Titulo') # configurando la columna titulo como indice para este DataFrame.
```

```
razao['Crédito'].sum()-razao['Débito'].sum()  #HACIENDO EL CALCULO DEL SALDO
```

##Carregando os dados referentes aos titulos e nomes dos fornecedores.

```
titulo=pd.DataFrame(pd.read_excel('F:/Conciliacao/CMA/TITULOS_CMA.xls', index_col='Titulo'))
```

```
titulo
```

```
#Visualizando os titulos e os fornecedores
titulo
```

```
titulo.sort_index
```

###Paso importante! HAciendo la juncion de los dos DataFrames razao e titulos

```
cma=pd.DataFrame(pd.merge(razao,titulo, on='Titulo'))
cma
```

```
fornecedor=pd.DataFrame(cma.groupby('Credor')['Débito','Crédito','Saldo'].sum())
fornecedor.to_excel('F:/Conciliacao/CMA/SAIDA_CMA/POR_FORNECEDOR.xlsx')
```

```
por_titulo=pd.DataFrame(cma.groupby(['Credor','Titulo'])['Histórico','Débito','Crédito','Saldo'].sum())
por_titulo.to_excel('F:/Conciliacao/CMA/SAIDA_CMA/POR_FORNECEDOR_E_TITULO.xlsx')
```

```
por_documento=pd.DataFrame(cma.groupby(['Credor','Titulo','Documento','Histórico','Data'])['Débito','Crédito','Saldo'].sum())
por_documento.to_excel('F:/Conciliacao/CMA/SAIDA_CMA/FORNECEDOR_TITULO_DOCUMENTO.xlsx')
```
