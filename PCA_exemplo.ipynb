```python
import sklearn
import mglearn
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
%matplotlib inline
```

```python
mglearn.plots.plot_pca_illustration()
```

```python
cancer=load_breast_cancer()
```

```python
cancer.feature_names.shape
```

```python
pca=PCA(n_components=2)
pca.fit(cancer.data)
```

```python
transformada=pca.transform(cancer.data)
transformada.shape
```

```python
cancer.data.shape
```

```python
transformada.shape
```

```python
mglearn.discrete_scatter(transformada[:,0],transformada[:,1],cancer.target)
plt.legend(cancer.target_names,loc='best')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
```

```python
from sklearn.preprocessing import MinMaxScaler
escala=MinMaxScaler()
escala.fit(cancer.data)
escalada=escala.transform(cancer.data)
pca.fit(escalada)
transformada=pca.transform(escalada)
mglearn.discrete_scatter(transformada[:,0],transformada[:,1],cancer.target)
plt.legend(cancer.target_names,loc='best')
plt.gca()
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
```
