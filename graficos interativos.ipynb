```python
df=pd.DataFrame(datos.groupby('date').sum())
df1=pd.DataFrame(df.resample('d')['price'].mean()).replace({np.nan:0.00})
```

```python
def filtro(x):
	c=0
	maior=[]
	minor=[]
	for value in x:
		if value==0:
			minor.append(value)
		else:
			maior.append(value)
			c+=1
	return maior
```
