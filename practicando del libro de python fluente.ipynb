```python
import pandas as pd
import numpy as np
```

```python
f=pd.DataFrame(np.random.random(50), columns=['Cod.'], index=range(50))
f
```

```python
def algo(x):
    for i in x:
        lice=x*np.random.randint(2,9)
        return lice
```

```python
h=algo(f['Cod.'])
f['B']=algo(f['Cod.'])
f
```

```python
from matplotlib import pyplot as plt
s=f[10:20]
s
```

```python
plt.scatter(s['Cod.'],np.arange(len(s)), label='Cod.', marker='o', color='red')
plt.scatter(sorted(s['B']),np.arange(len(s)), label='B', marker='x', color='green')
plt.legend(loc=1)
plt.title('Tallas que rompen tallas', fontsize=24)
plt.xlabel('')
```
