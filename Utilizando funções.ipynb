```python
data=np.random.randn(2500,4)
Dataframe=pd.DataFrame(data,columns=['a','b','c','d'])
Dataframe.index.names=['order']
Dataframe.columns.names=['reagentes']
Dataframe
```

```python
states=['Alabama','Georgia','Georgia','georgia','FlOrida', 'south carolina##','west virginia']
```

```python
import re
```

```python
def clean_strings(strings):
    result=[]
    for value in strings:
        value=value.strip()
        value=re.sub('[!#?]', '',value)
        value=value.title()
        result.append(value)
    return result
```

```python
estados=clean_strings(states)
```

```python
estados
```

```python
def Classification(b,opcoes):
    c=0
    a=len(b)
    result=[]
    while c<a:
        for i in range(a):
            result.append(random.choice(opcoes))
        return result
    c+=1
```

```python
Dataframe['classif']=Classification(Dataframe,choice)
```
