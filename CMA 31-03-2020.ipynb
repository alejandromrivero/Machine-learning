```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import seaborn as sb
import datetime
```

```python
razao=pd.read_excel('F:/relatorio (27).xlsx')
```

```python
razao=pd.DataFrame(razao)
razao['Titulo']=pd.to_numeric(razao['Titulo'])
razao['Data']=pd.to_datetime(razao['Data'])
razao['Débito']=razao['Débito'].replace({np.nan:0.00})
razao['Crédito']=razao['Crédito'].replace({np.nan:0.00})
razao
```

```python
titulo=pd.DataFrame(pd.read_excel('F:/relatorio (28).xlsx'))
titulo['Dt.CT/OC']=pd.to_datetime(titulo['Dt.CT/OC'])
titulo['Emis.NF']=pd.to_datetime(titulo['Emis.NF'])
titulo['Cadastro']=pd.to_datetime(titulo['Cadastro'])
titulo['Vencto']=pd.to_datetime(titulo['Vencto'])
titulo['Titulo']=pd.to_numeric(titulo['Titulo'])
titulo
```

```python
titulo.dtypes
```

```python
cma_janeiro=pd.DataFrame(pd.merge(razao, titulo, on='Titulo'))
```

```python
cma_janeiro=cma_janeiro.set_index(['Titulo','Data'])
```

```python
cma_janeiro.dtypes
```

```python
cma_janeiro['Saldo']=cma_janeiro['Crédito']-cma_janeiro['Débito']
```

```python
cma_janeiro
```

```python
cma_janeiro.groupby(['Credor','Titulo'])['Débito','Crédito','Saldo'].sum()
```

```python
por_fornecedor=pd.DataFrame(cma_janeiro.groupby(['Credor','Data'])['Débito','Crédito','Saldo'].sum())
por_fornecedor.to_excel('F:/por_fornecedor.xlsx')
```

```python
por_titulo=pd.DataFrame(cma_janeiro.groupby(['Credor','Titulo'])['Débito','Crédito','Saldo'].sum())
por_titulo.to_excel('F:/por_titulo.xlsx')
```

```python
por_documento=pd.DataFrame(cma_janeiro.groupby(['Credor','Titulo','Documento'])['Débito','Crédito','Saldo'].sum())
por_documento.to_excel('F:/por_documento.xlsx')
```
