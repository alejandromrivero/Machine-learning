```python
import folium
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import selenium
```

```python
# Mudando o tipo de gráfico
m = folium.Map(location=[-20.22912, -40.27991],
               zoom_start=20, tiles= 'Stamen Toner')
folium.Marker(location=[-20.22912, -40.27991],
              popup='Minha casa',
              icon=folium.Icon(color='green', icon='info-sign',angle=90)
).add_to(m)
m.save('F:/andre-carloni.html')
```

```python
# Primeiro criamos o objeto m para o mapa
m = folium.Map(location=[-20.22912, -40.27991], zoom_start=18)
# Depois adicionamos o marcador.
folium.CircleMarker(location=[-20.22912, -40.27991],
                radius=50,
                popup='Casa de Fabiana e Alejandro',
                icon=folium.Icon(color='red', icon='info-sign')
).add_to(m)

m
```

```python
# Primeiro criamos o objeto m para o mapa
m = folium.Map(location=[-20.22912, -40.27991], zoom_start=70)# Adicionando o marcador circular
folium.CircleMarker(
    location=[-20.22912, -40.27991],
    radius=50,
    popup='Fabiana e Alejandro',
    color='blue',
    fill=True,
    fill_color='lightblue'
).add_to(m)

m
m.save('F:/mapa.html')
```

```python
# Criando o objeto m para o mapa
m = folium.Map(location=[-20.22912, -40.27991], zoom_start=14)
df=m.add_child(folium.ClickForMarker(popup='Adicionado pelo Usuario'))
df
```
