```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import plotly.offline as py
import plotly.graph_objs as go
```

###Cargando razao

```python
razao=pd.read_excel('C:/Users/User/Desktop/CMA versão 3/razao.xlsx')
```

```python
razao=pd.DataFrame(razao)
razao['Data']=pd.to_datetime(razao['Data'])
razao['Débito']=razao['Débito'].replace({np.nan:0.00})
razao['Crédito']=razao['Crédito'].replace({np.nan:0.00})
razao['Saldo']=razao['Crédito']-razao['Débito']
```

```python
razao['Crédito'].sum()-razao['Débito'].sum()
```

###Cargando titulo

```python
titulo=pd.DataFrame(pd.read_excel('C:/Users/User/Desktop/CMA versão 3/titulo.xlsx'))
```

```python
titulo.dtypes
```

```python
cma=pd.DataFrame(pd.merge(razao,titulo, on='Titulo'))
```

```python
cma=cma.set_index('Titulo')
cma
```

```python
cma.count()
```

```python
credor=pd.DataFrame(cma.groupby(['Credor'])['Débito','Crédito','Saldo'].sum())
credor.to_excel('C:/Users/User/Desktop/CMA versão 3/Relatorio01.xlsx')
```

```python
credor_titulo=pd.DataFrame(cma.groupby(['Credor','Titulo'])['Débito','Crédito','Saldo'].sum())
credor_titulo.to_excel('C:/Users/User/Desktop/CMA versão 3/Relatorio02.xlsx')
```

```python
credor_titulo_doc=pd.DataFrame(cma.groupby(['Credor','Titulo','Documento'])['Débito','Crédito','Saldo'].sum())
credor_titulo_doc.to_excel('C:/Users/User/Desktop/CMA versão 3/Relatorio03.xlsx')
```
