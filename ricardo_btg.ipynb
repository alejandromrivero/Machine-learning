```python
import numpy as np
import pandas as pd
```

```python
proventos_btg=pd.DataFrame(pd.read_excel('F:/resumo_cipriano_BTG.xlsx'))
proventos_btg['Valor_Bruto']=pd.to_numeric(proventos_btg['Valor_Bruto'])
proventos_btg['IR']=pd.to_numeric(proventos_btg['IR'])
proventos_btg['Valor_Liquido']=pd.to_numeric(proventos_btg['Valor_Liquido'])
proventos_btg
```

```python
proventos_btg_grouped=proventos_btg.groupby(['Ativo','Tipo_Evento'])['Valor_Bruto','IR','Valor_Liquido'].sum()
proventos_btg_grouped.to_excel('F:/resumo_cipriano_BTG_grouped.xlsx')
```
