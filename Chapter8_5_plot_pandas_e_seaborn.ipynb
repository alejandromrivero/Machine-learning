```python
data2=macro[['cpi','m1','tbilrate','unemp']]
```

```python
trans_data2=np.log(data2).diff().dropna()
```

```python
trans_data2[-5:]
```

```python
sns.regplot('m1','unemp',data=trans_data2, color='k')
plt.plot(figsize=(10,10))
plt.title('Changes in log m1 versus log unemp')
plt.savefig('F:\dumps\Saida de regplot do seaborn.jpg')
```
